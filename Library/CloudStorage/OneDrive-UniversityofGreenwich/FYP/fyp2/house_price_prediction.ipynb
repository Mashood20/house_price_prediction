```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Bengaluru_House_Data 2.csv")
df.shape
```

```python
df.head()
```

```python
df.groupby('area_type').agg('count')
```

```python
#deleting availability as seem to be of no use
df2  = df.drop(['availability', 'society'], axis='columns')
df2.head()
```

```python
#cleaning start from here

#checking which column have how many null values 
df2.isnull().sum()

#imputing value zero where balcony is null
df2['balcony'] =  df2['balcony'].fillna(0.0)
df3 = df2.dropna()
df3.isnull().sum()
```

```python
#size has different types of format so gonna makesure they have same format

#checking different types of values we have
df3['size'].unique()
# creating a new column
df3['BHK'] = df3['size'].apply(lambda x: int(x.split(' ')[0]))

df3.head()
```

```python
#checking if we dont have different formats
df3['BHK'].unique()

#ok we found some houses with too many bedroom gonna check them further
df3[df3.BHK>20]

#so as guessed found some absurd listin, so gonna drop that
df4 = df3[df3['BHK']<=20]

```

```python
#now lets check the total_sqft
df4['total_sqft'].unique()
# found different format as some are given in range so what iam gonna do is take average of the range

#first lets check if it is float for that iam writing is_float function which will filter out the values which can be converted to float.
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True

#iam negating the function using ~ to select only which cannot be converted
df4[~df4['total_sqft'].apply(is_float)]
```

```python
#here iam writing function to take average of these ranges
def taking_avg(x):
    tokens =  x.split('-')
    if len(tokens)==2:
        return (float(tokens[0])+float(tokens[0]))/2
    try:
        return float(x)
    except:
        return None

#applying the function to column

df5 = df4.copy()

df5['total_sqft'] = df5['total_sqft'].apply(taking_avg)
df5.head()
```

```python
#checking if all went good 
df5.loc[122]

#dropping size as we already created a seprate column

df6 = df5.drop(['size'], axis=1)
df6.head()
```

```python
#here iam doing a feature engineering by adding the feature price per sqft which will also help to check if there is outlier

# multiplying by 100000 as price is in hundred thousands
df6['price_per_sqft'] = df6['price']*100000/ df6['total_sqft']
df6.head()
```
